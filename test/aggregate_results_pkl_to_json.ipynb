```python
import pickle
import numpy as np
import gzip
```

```python
results_name = 'unconstrained_rerun_initial_datasets_results_1000bootstraps_100fits'
```

```python
with open(f'/data/ross/assay_calibration/{results_name}.pkl','rb') as f:
    results = pickle.load(f)
```

```python
import json
from sklearn.cluster import KMeans

class NumpyEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        if isinstance(obj, (np.floating, np.float64, np.float32)):
            return float(obj)
        if isinstance(obj, (np.integer, np.int64, np.int32)):
            return int(obj)
        if isinstance(obj, np.bool_):
            return bool(obj)
        if isinstance(obj, KMeans):
            # Extract relevant KMeans info
            return {
                'type': 'KMeans',
                'n_clusters': int(obj.n_clusters),
                'cluster_centers': obj.cluster_centers_.tolist() if hasattr(obj, 'cluster_centers_') else None,
                'labels': obj.labels_.tolist() if hasattr(obj, 'labels_') else None,
                'inertia': float(obj.inertia_) if hasattr(obj, 'inertia_') else None
            }
        return super().default(obj)


with gzip.open(f'/data/ross/assay_calibration/{results_name}.json.gz', 'wt', encoding='utf-8') as f:
    json.dump(results, f, cls=NumpyEncoder, indent=2)

print("Converted pickle to JSON successfully")
```
